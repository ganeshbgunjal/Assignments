**Que:- 12** Write a program that calculates the factorial of a number input by the user using a while loop.


```python
factorial = 1
num = int(input('Enter num:'))
for i in range(1, num+1):
    factorial = factorial * i
print(f'Factorial of {num} is {factorial}.')
```

**Que:-13** Write a program that calculates the sum of the digits of a number input by the user using a while loop.
 

```python
num = int(input('Enter num:'))
sum = 0

while num > 0:
    digit = num % 10
    sum = sum + digit
    num//=10   #its for removing last digit. its integer division.

print(f'The sum of all digits is: {sum}')
    
```

**Que:- 14** Write a program that checks if a number input by the user is a prime number using a for loop.


```python
num = int(input('Enter num:'))
if num <2:
        print(f'{num} is not a prime number.')
else:
    for i in range(2,num):
        if num % i == 0:
            print(f'{num} is not a prime number.')
            break
    else:
        print(f'{num} is a prime number.')
```
