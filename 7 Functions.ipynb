### Assignment 1: Fibonacci Sequence with Memoization

Define a recursive function to calculate the nth Fibonacci number using memoization. Test the function with different inputs.  


**The memo parameter in your fibonacci function is an example of memoization, which is an optimization technique to store the results of expensive function calls and return the cached result when the same inputs occur again.**

```python
def fibonacci(n,memo={}):
    if n in memo:
        return memo[n]
    if n<=1:
        return n
    memo[n]= fibonacci(n-1,memo) + fibonacci(n-2,memo)
    return memo[n]
```

```python
print(fibonacci(10))
print(fibonacci(15))
```
