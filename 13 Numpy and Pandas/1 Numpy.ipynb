2. Create a NumPy array of shape (4, 4) with values from 1 to 16. Replace the diagonal elements with 0.

```python
array = np.arange(1,17).reshape((4,4))
print('--- Original Array ---')
print(array)
print('--- After changing diagonal elements ---')
np.fill_diagonal(array,0)
print(array)
```

2. Create a NumPy array of shape (5, 5) with random integers. Extract the elements on the border.

```python
array = np.random.randint(1,21,size=((5,5)))
print('--- Original Array ---')
print(array)
border_elements = np.concatenate((array[0,:],array[-1,:],array[1:-1,0],array[1:-1,-1]))
print('--- Border Elements ---')
print(border_elements)
```

### Assignment 3: Array Operations

1. Create two NumPy arrays of shape (3, 4) filled with random integers. Perform element-wise addition, subtraction, multiplication, and division.

```python
array1 = np.random.randint(1,11,size=(3,4))
array2 = np.random.randint(1,11,size=(3,4))

print('-- Array1 --')
print(array1)
print('-- Array2 --')
print(array2)

addition = array1+array2
substraction = array1-array2
multiplication = array1*array2
division = array1/array2

print('-- Elementwise addition --')
print(addition)
print('-- Elementwise substraction --')
print(substraction)
print('-- Elementwise multiplication --')
print(multiplication)
print('-- Elementwise division --')
print(division)
```

2. Create a NumPy array of shape (4, 4) with values from 1 to 16. Compute the row-wise and column-wise sum.

```python
array = np.arange(1,17).reshape((4,4))
print('-- Original Array --')
print(array)

row_sum = np.sum(array,axis=1)
col_cum = np.sum(array,axis=0)

print('-- Rowwise sum --')
print(row_sum)
print('-- Colwise sum --')
print(col_cum)
```

### Assignment 4: Statistical Operations

1. Create a NumPy array of shape (5, 5) filled with random integers. Compute the mean, median, standard deviation, and variance of the array.

```python
array = np.random.randint(1,21,size=(5,5))
print('-- Original Array --')
print(array)

mean = np.mean(array)
median = np.median(array)
std_dev = np.std(array)
var = np.var(array)

print('-- Mean --')
print(mean)
print('-- Median --')
print(median)
print('-- Standard Deviation --')
print(std_dev)
print('-- Variance --')
print(var)
```

Mean: it calculates the mean of all 25 elements in the array — not row-wise or column-wise, just a single average of everything.  
Median: The median is the middle value in a sorted list.  
- If there’s an odd number of elements → it's the center one.
- If there’s an even number → it's the average of the two middle values.  
Standard deviation measures how spread out the numbers are from the mean.
