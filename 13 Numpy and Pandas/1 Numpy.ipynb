### Assignment 4: Statistical Operations

1. Create a NumPy array of shape (5, 5) filled with random integers. Compute the mean, median, standard deviation, and variance of the array.

```python
array = np.random.randint(1,21,size=(5,5))
print('-- Original Array --')
print(array)

mean = np.mean(array)
median = np.median(array)
std_dev = np.std(array)
var = np.var(array)

print('-- Mean --')
print(mean)
print('-- Median --')
print(median)
print('-- Standard Deviation --')
print(std_dev)
print('-- Variance --')
print(var)
```

Mean: it calculates the mean of all 25 elements in the array — not row-wise or column-wise, just a single average of everything.  
Median: The median is the middle value in a sorted list.  
- If there’s an odd number of elements → it's the center one.
- If there’s an even number → it's the average of the two middle values.  
Standard deviation measures how spread out the numbers are from the mean.



2. Create a NumPy array of shape (3, 3) with values from 1 to 9. Normalize the array (i.e., scale the values to have a mean of 0 and a standard deviation of 1).

```python
# Create the original array

array = np.arange(1,10).reshape((3,3))
print('-- Original Array --')
print(array)

#normalize the array
mean = np.mean(array)
std_dev = np.std(array)
normalized_arr = (array-mean)/std_dev

print('-- Normalised array --')
print(normalized_arr)
```

### Assignment 5: Broadcasting

1. Create a NumPy array of shape (3, 3) filled with random integers. Add a 1D array of shape (3,) to each row of the 2D array using broadcasting.

```python
#create a original array
array = np.random.randint(1,11, size=(3,3))
row_array = np.random.randint(1,11,size=(3,))

print('-- Original array --')
print(array)
print('-- 1D array --')
print(row_array)

#Add the 1D array to each array using broadcasting
result = array + row_array
print('-- Resulted array --')
print(result)
```

### Assignment 6: Linear Algebra

1. Create a NumPy array of shape (3, 3) representing a matrix. Compute its determinant, inverse, and eigenvalues.

```python
#create the matrix with the size (3,3)
matrix = np.random.randint(1,11,size=(3,3))
print('-- Matrix --')
print(matrix)

#calculate its determinant
determinant = np.linalg.det(matrix)
print(f'Determinant: {determinant}')

#calculate the Inverse
inverse = np.linalg.inv(matrix)
print(f'Inverse: {inverse}')

#calculate the Eigen values
eigenvalues = np.linalg.eigvals(matrix)
print(f'Eigen values: {eigenvalues}')
```

### Assignment 7: Advanced Array Manipulation

1. Create a NumPy array of shape (3, 3) with values from 1 to 9. Reshape the array to shape (1, 9) and then to shape (9, 1).

```python
#create the array of (3,3)
array = np.arange(1,10).reshape(3,3)
print('Original array')
print(array)

#reshape this array to (1,9)
reshaped_array1 = array.reshape(1,9)
print('Reshaped array 1')
print(reshaped_array1)

#again reshape to array(9,1)
reshaped_array2 = reshaped_array1.reshape(9,1)
print('Reshaped array 2')
print(reshaped_array2)
```

2. Create a NumPy array of shape (4, 4) filled with random integers. Use boolean indexing to set all elements greater than 10 to 10.

```python
# create a numpy array of (4,4) with random integers.
array = np.random.randint(1,21,size=(4,4))
print('Original Array')
print(array)

#use boolean indexing to set elements >10 to 10
array[array >10] = 10
print('Modified Array')
print(array)
```
