### Assignment 1: Single Inheritance Basic

Create a base class named `Animal` with attributes `name` and `species`. Create a derived class named `Dog` that inherits from `Animal` and adds an attribute `breed`. Create an object of the `Dog` class and print its attributes.

```python
class Animal:
    def __init__(self,name,species):
        self.name = name
        self.species = species

class Dog(Animal):
    def __init__(self,name,species,breed):
        super().__init__(name,species)
        self.breed = breed

#create an object of dog:
dog = Dog('Lyka','Dog','German shephard')

#print attributes:
print(f'Name: {dog.name}')
print(f'Species: {dog.species}')
print(f'Breed: {dog.breed}')
```

### Assignment 2: Method Overriding in Single Inheritance

In the `Dog` class, override the `__str__` method to return a string representation of the object. Create an object of the class and print it.


```python
class Animal:
    def __init__(self,name,species):
        self.name = name
        self.species = species

class Dog(Animal):
    def __init__(self,name,species,breed):
        super().__init__(name,species)
        self.breed = breed

    def __str__(self):
        return f'{self.name} is a {self.breed} Dog!'

#create an object of dog:
dog = Dog('Lyka','Dog','German shephard')

#print __str__  attributes:
print(dog)
```

### Assignment 3: Single Inheritance with Additional Methods

In the `Dog` class, add a method named `bark` that prints a barking sound. Create an object of the class and call the method.

```python
class Animal:
    def __init__(self,name,species):
        self.name = name
        self.species = species

class Dog(Animal):
    def __init__(self,name,species,breed):
        super().__init__(name,species)
        self.breed = breed

    def __str__(self):
        return f'{self.name} is a {self.breed} Dog!'

    def bark(self):
        return f'{self.name} barks like bho bho'

#create an object of dog:
dog = Dog('Lyka','Dog','German shephard')

#print __str__  attributes:
print(dog)
print(dog.bark())
```

### Assignment 4: Multiple Inheritance Basic

Create a base class named `Walker` with a method `walk` that prints a walking message. Create another base class named `Runner` with a method `run` that prints a running message. Create a derived class named `Athlete` that inherits from both `Walker` and `Runner`. Create an object of the `Athlete` class and call both methods.

```python
class Walker:
    def walk(self):
        print('The athelete is walking!')
class Runner:
    def run(self):
        print('The athelete is running')
class Athlete(Walker,Runner):
    pass

athlete1 = Athlete()
athlete1.run()
athlete1.walk()
```

### Assignment 5: Method Resolution Order (MRO) in Multiple Inheritance

In the `Athlete` class, override the `walk` method to print a different message. Create an object of the class and call the `walk` method. Use the `super()` function to call the `walk` method of the `Walker` class.

```python
class Walker:
    def walk(self):
        print('The athelete is walking')

class Runner:
    def run(self):
        print('The athlete is running')

class Athlete(Walker,Runner):
    def walk(self):
        print('The athlete is warming up slowly')
        super().walk()

athlete = Athlete()
athlete.walk()
athlete.run()
```

### Assignment 6: Multiple Inheritance with Additional Attributes

In the `Athlete` class, add an attribute `training_hours` and a method `train` that prints the training hours. Create an object of the class and call the method.

```python
class Walker:
    def walk(self):
        print('The athelete is walking')

class Runner:
    def run(self):
        print('The athelete is running')

class Athlete(Walker,Runner):
    def __init__(self,training_hrs):
        self.training_hrs = training_hrs
        super().walk()

    def train(self):
        print(f'The athlete trains {self.training_hrs} hrs every day')

athlete = Athlete(16)
athlete.run()
athlete.walk()
athlete.train()
```

### Assignment 7: Diamond Problem in Multiple Inheritance

Create a class named `A` with a method `show` that prints a message. Create two derived classes `B` and `C` that inherit from `A` and override the `show` method. Create a class `D` that inherits from both `B` and `C`. Create an object of the `D` class and call the `show` method. Observe the method resolution order.

```python
class A:
    def show(self):
        print('Class A show method')

class B(A):
    def show(self):
        print('Class B show method')

class C(A):
    def show(self):
        print('Class C show method')

class D(B,C):
    pass

d = D()
d.show()

# Display the Method Resolution Order (MRO)
print(D.__mro__)
```

### Assignment 8: Using `super()` in Single Inheritance

Create a base class named `Shape` with an attribute `color`. Create a derived class named `Circle` that inherits from `Shape` and adds an attribute `radius`. Use the `super()` function to initialize the attributes. Create an object of the `Circle` class and print its attributes.

```python
class Shape:
    def __init__(self,color):
        self.color = color

class Circle(Shape):
    def __init__(self,color,radius):
        super().__init__(color)
        self.radius = radius

circle = Circle('red',5)
print(f'Color: {circle.color}')
print(f'Radius: {circle.radius}')
```

### Assignment 9: Using `super()` in Multiple Inheritance

Create a class named `Person` with an attribute `name`. Create a class named `Employee` with an attribute `employee_id`. Create a derived class `Manager` that inherits from both `Person` and `Employee`. Use the `super()` function to initialize the attributes. Create an object of the `Manager` class and print its attributes.

```python
class Person:
    def __init__(self,name):
        self.name = name
        print("Person initialized")

class Employee:
    def __init__(self,emp_id):
        self.emp_id = emp_id
        print("Employee initialized")

class Manager(Person,Employee):
    def __init__(self,name,emp_id):
        super().__init__(name)
        self.emp_id = emp_id

manager = Manager('Ganesh',101)
print(manager.name)
print(manager.emp_id)

# Show Method Resolution Order
print(Manager.__mro__)
    
```
