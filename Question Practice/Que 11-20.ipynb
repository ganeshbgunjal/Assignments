### Q11: Write a Python program to print numbers from 1 to 10 using a while loop.

Explanation:
Initialize i = 1.
Use while i <= 10:
- print i
- increment i by 1 each time.
Loop stops when condition becomes false.

```python
i=1
while i <=10:
    print(i)
    i = i+1
    
```

### Q12: Write a Python program to print the multiplication table of a given number n (from 1 to 10).

Explanation:    <br>
Take an integer n.      <br>
Use a for loop from 1 to 10.      <br>
On each iteration, multiply n by i and print “n x i = result”.

```python
num = int(input('Enter num:'))
for i in range(1,11):
    print(f'{num} * {i} = {i*num}')
```
