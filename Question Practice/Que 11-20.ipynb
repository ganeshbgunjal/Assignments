### Q17: Write a Python program to find the factorial of a number using recursion.

Explanation:
Recursive definition of factorial:
- factorial(0) = 1
- factorial(1) = 1
- factorial(n) = n * factorial(n - 1) for n > 1
We call the function recursively until the base case.

```python
num = int(input('Enter num:'))
def factorial(num):
    if num ==0 or num==1:
        return 1
    else:
        return num * factorial(num-1)
print(f'Factorial of {num} is {factorial(num)}')
```

### Q18: Write a Python program to reverse the digits of a number using a loop.

Explanation:
We can repeatedly take the last digit using n % 10, and build a new number.
Steps:
1) Initialize rev = 0.
2) While n > 0:
   - digit = n % 10
   - rev = rev * 10 + digit
   - n = n // 10 (integer division)
3) rev will be the reversed number.

```python
num = input('Enter num:')
rev= ""

for digit in num:
    rev = digit + rev
print(f'Reversed: {rev}')
```

```python
num = int(input('Enter num:'))
rev = 0

while num >0:
    digit = num %10    #get last digit
    rev = rev * 10 + digit    #build reversed number
    num = num // 10    #remove last digit
print(f'Reversed Number: {rev}')
    
```

### Q19: Write a Python program to check if a number is prime.

Explanation:
A prime number is greater than 1 and has no divisors other than 1 and itself.
Algorithm:
1) If n <= 1 → not prime.
2) Check divisors from 2 to sqrt(n). If any divides n, it is not prime.
3) If no divisor found, it's prime.

```python
num = int(input('Enter num:'))

if num <2:
    print(f'{num} is not a prime number')
else:
    for i in range(2,num):
        if num % i ==0:
            print(f'{num} is not a prime number')
            break
    else:
        print(f'{num} is a prime number')
```

### Q20: Write a Python program to count the number of digits in a given number.

Explanation:    <br>
We can repeatedly divide the number by 10 (integer division) until it becomes 0.   <br>
Each division step removes one digit. We count how many steps we perform.   <br>
Edge case: if number is 0, it has 1 digit.

```python
num = int(input('Enter number:'))

count = 0

while num >0:
    num = num//10
    count+=1
print(f'Total digits are:{count}')
```

```python
num = input('Enter num:')
print(f'Total digits in number: {len(num)}')
```

### Write a Python program to find the greatest of n numbers entered by the user.

Explanation:
We can:
1) Ask user how many numbers they will enter (n).
2) Loop n times, read each number, and keep track of the current maximum.
3) After the loop, print the maximum.

```python
n = int(input("How many numbers will you enter? "))

largest = None   # or float('-inf')

for i in range(n):
    num = int(input(f"Enter number {i+1}: "))
    
    if largest is None or num > largest:
        largest = num

print(f"The greatest number is: {largest}")
```
