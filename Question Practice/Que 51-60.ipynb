### Q51: Write a Python function that counts how many prime numbers are between 1 and n.

Explanation:   <br>
We loop from 2 to n and check if each number is prime.    <br>    
We reuse the logic of checking prime (divisible only by 1 and itself).  <br>
For each prime number, we increment the counter.
Finally, we return the total count.

```python
def check_prime(n):
    count = 0

    for num in range(2,n+1):
        is_prime = True

        for i in range(2, int(num ** 0.5) + 1):
            if num % i == 0:
                is_prime = False
                break
        if is_prime:
            count+=1

    return count


n =20
print(f'Prime numbers between 1 to {n} are {check_prime(n)}')
        
    
```

### Q52: Write a Python program to generate the first n Fibonacci numbers using a loop.

Explanation:    <br>
The Fibonacci sequence:     <br>
0, 1, 1, 2, 3, 5, 8, ...     <br>
Every number is the sum of the previous two numbers.   <br>
We start with a = 0 and b = 1.    <br>
On each iteration, we append a, then update a and b.

```python
def fibonacci(n):
    fib_series = []

    a, b = 0, 1

    for _ in range(n):
        fib_series.append(a)
        a, b = b, a + b

    return fib_series


# Example usage
n = 10
print("First", n, "Fibonacci numbers:", fibonacci(n))
```

### Q53: Write a Python recursive function to generate the nth Fibonacci number.

Explanation:    <br>
Definition:     <br>
fib(0) = 0      <br>
fib(1) = 1     <br>
fib(n) = fib(n-1) + fib(n-2)   <br> 
This definition fits recursion perfectly.   <br>
Because recursion is slow, use it only for learning.

```python
def fibonacci(n):
    if n ==0:
        return 0
    elif n==1:
        return 1
    else:
        return fibonacci(n-1)+fibonacci(n-2)

n=7
print(f'Fibonnaci of {n} is {fibonacci(n)}')
    
```

### Q54: Write a Python function that takes a string and returns the number of words.

Explanation:    <br>
Words are separated by spaces.    <br>
We use split() to break the sentence into words.   <br>
Then we return the length of the resulting list.

```python
def count_words():
    str1= input('Enter string:')
    str1 = str1.split(" ")
    print(f'Total words in the strings are: {len(str1)}')

count_words()
```
