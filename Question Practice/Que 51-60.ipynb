### Q51: Write a Python function that counts how many prime numbers are between 1 and n.

Explanation:   <br>
We loop from 2 to n and check if each number is prime.    <br>    
We reuse the logic of checking prime (divisible only by 1 and itself).  <br>
For each prime number, we increment the counter.
Finally, we return the total count.

```python
def check_prime(n):
    count = 0

    for num in range(2,n+1):
        is_prime = True

        for i in range(2, int(num ** 0.5) + 1):
            if num % i == 0:
                is_prime = False
                break
        if is_prime:
            count+=1

    return count


n =20
print(f'Prime numbers between 1 to {n} are {check_prime(n)}')
        
    
```

### Q52: Write a Python program to generate the first n Fibonacci numbers using a loop.

Explanation:    <br>
The Fibonacci sequence:     <br>
0, 1, 1, 2, 3, 5, 8, ...     <br>
Every number is the sum of the previous two numbers.   <br>
We start with a = 0 and b = 1.    <br>
On each iteration, we append a, then update a and b.

```python
def fibonacci(n):
    fib_series = []
    
    a,b = 0,1

    for _ in range(n):
        fib_series.append(a)
        a,b = b, a+b
    return fib_series

n=10
print(f'First {n} Fibonacci series is: {fibonacci(n)}')
```

### Q53: Write a Python recursive function to generate the nth Fibonacci number.

Explanation:    <br>
Definition:     <br>
fib(0) = 0      <br>
fib(1) = 1     <br>
fib(n) = fib(n-1) + fib(n-2)   <br> 
This definition fits recursion perfectly.   <br>
Because recursion is slow, use it only for learning.

```python
def fibonacci(n):
    if n ==0:
        return 0
    elif n==1:
        return 1
    else:
        return fibonacci(n-1)+fibonacci(n-2)

n=7
print(f'Fibonnaci of {n} is {fibonacci(n)}')
    
```

### Q56: Write a Python function that takes a list of numbers and returns the median.

Explanation:   <br>
Median:   <br>
- Sort the list.  <br>
- If length is odd, return middle element.   <br>
- If even, return average of the two middle elements.  <br>
We find indexes accordingly.

```python
def get_median(nums):
    nums = sorted(nums)
    n= len(nums)

    if n ==0:
        return None

    mid = n//2
    if n%2 != 0:
        return nums[mid]
    else:
        return (nums[mid - 1] + nums[mid]) / 2


numbers = [7, 3, 5, 1, 9,4]
print("Median:", get_median(numbers))
    
```

### Q57: Write a Python program to find the GCD (Greatest Common Divisor) of two numbers.

Explanation:    <br>
GCD is the largest number that divides both a and b.  <br>
We can use the Euclidean algorithm:  <br>
gcd(a, b) = gcd(b, a % b)  <br>
When b becomes 0, a is the GCD.

```python
def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a


# Example usage
a = 48
b = 18
print(f"GCD of {a} and {b} is:", gcd(a, b))
```

```python
def find_gcd(a,b):
    while b!=0:
        a,b = b,a%b
    return a

a = 48
b = 18
print(f'GCS if {a} and {b} is {find_gcd(a,b)}')
```

### Q58: Write a Python function to compute the LCM (Least Common Multiple) of two numbers.

Explanation:   <br>
Relation between GCD and LCM:   <br>
LCM(a, b) = (a * b) // GCD(a, b)   <br>
We reuse the gcd() function and apply the above formula.

```python
def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a


def lcm(a, b):
    return (a * b) // gcd(a, b)


# Example usage
a = 48
b = 18
print(f"GCD of {a} and {b} is:", gcd(a, b))
print(f"LCM of {a} and {b} is:", lcm(a, b))
```

### Q59: Write a Python function to check if two strings are anagrams.

Explanation:   <br>
Two strings are anagrams if they contain the same characters in any order.    <br>
Approach:   <br>
- Sort both strings and compare results.   <br>
If sorted versions are equal, they are anagrams.

```python
def check_anagrams(s1,s2):
    s1 = s1.replace(" ","").lower()
    s2 = s2.replace(" ","").lower()
    return sorted(s1) == sorted(s2)

s1 = 'listEn'
s2 = 'silent'
print(f'Strings are Anagrams? {check_anagrams(s1,s2)}')
```

### Q60: Write a Python program to find all factors of a given number.

Explanation:    <br>
A factor divides the number without remainder.   <br>
We loop from 1 to n.   <br>
If n % i == 0, then i is a factor.  <br>
We append all such i to the result list.

```python
def find_factors(n):
    factors = []
    for i in range(1,n+1):
        if n % i == 0:
            factors.append(i)
    print(factors)

find_factors(28)
```
