### Q41: Write a Python function to compute the factorial of a number using recursion.

Explanation:    <br>
Factorial definition:    <br>
- factorial(0) = 1
- factorial(1) = 1
- factorial(n) = n * factorial(n-1), for n > 1.   <br>
We implement this directly using a recursive function.

```python
def factorial(num):
    if num ==0 or num == 1:
        return 1
    else:
        return num * factorial(num-1)

num = int(input('Enter num:'))
print(f'Factorial of {num} is {factorial(num)}')
```

### Q42: Write a Python function that returns the maximum of three numbers.
 
Explanation:  <br>
We take three arguments a, b, c.    <br>
We initialize maximum with a and then compare b and c with it.   <br>
Finally, we return the maximum value.

```python
def max_number(a,b,c):
    if a>b and a>c:
        return a
    elif b>a and b>c:
        return b
    else:
        return c

result = max_number(11,33,22)

print(f'Maximum number is {result}')
```

### Q43: Write a Python function to check if a number is prime.

Explanation:
A prime number is greater than 1 and has no divisors other than 1 and itself.
We:
1) Return False if n <= 1.
2) Loop from 2 to sqrt(n) and check if n is divisible by any i.
3) If divisible, return False; else return True.

```python
def prime(n):
    if n <=1:
        return False
    for i in range(2,int(n**0.5)+1):
        if n % i ==0:
            return False
    return True

num = int(input('Enter num:'))
if prime(num):
    print(f'{num} is prime')
else:
    print(f'{num} is not prime')
```

### Q44: Write a Python function that takes a list of numbers and returns their average.

Explanation:     <br>
Average (mean) = sum of numbers / count of numbers.    <br>
If list is empty, we can return 0 or handle as a special case.    <br>
We use sum() and len() to calculate the average.

```python
def average(numbers):
    if len(numbers)==0:
        return 0
    return sum(numbers)/len(numbers)

total = average([1,2,3,4,5,6,7])
print(f'Average is {total}')
```

### Q45: Write a Python program that uses a list comprehension to create a list of squares from 1 to 10.

List comprehensions provide a compact syntax to create lists.    <br>

### Q47: Write a Python program that uses lambda and map() to square all numbers in a list.

Explanation:    <br>
lambda defines anonymous functions.   <br>
map() applies a function to each element of an iterable.   <br>
We define a lambda x: x**2 and map it over the list, then convert the result to a list.

```python
numbers = [1,2,3,4,5,6]
sq = list(map(lambda x : x**2,numbers))
print(sq)
```

### Q48: Write a Python program that uses filter() to keep only positive numbers from a list.

Explanation:   <br>
filter(function, iterable) keeps only elements for which function returns True.   <br>
We use a lambda that checks x > 0.    <br>

```python
numbers = [-10, 5, -3, 7, 0, -2, 9]

positive_num = list(filter(lambda x:x>0,numbers))
print(positive_num)
```

### Q49: Write a Python function to flatten a 2D list into a 1D list.

Explanation:    <br>
We have a list of lists (2D).   <br>
We loop through each row (inner list) and then each element in the row.   <br>
We append all elements to a new list, effectively flattening it.


```python
def flat_2d_list(matrix):
    return [item for sublist in matrix for item in sublist ]

data = [[1, 2, 3], [4, 5], [6, 7, 8]]
print(flat_2d_list(data))
```

### 50: Write a Python function to get unique elements from a list while preserving order.

Explanation:    <br>
We use a set to remember which elements we have already seen.   <br>
We iterate through the list; if an element is not in the set, we add it to both the set and the result list.   <br>
This removes duplicates but preserves the first occurrence order.

```python
def unique_preserve_order(items):
    seen = set()
    result = []

    for item in items:
        if item not in seen:
            seen.add(item)
            result.append(item)
    return result

numbers = [1, 2, 2, 3, 4, 3, 5, 1]
print(unique_preserve_order(numbers))
    
```
