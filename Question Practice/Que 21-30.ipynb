### Q21: Write a Python program to count the number of vowels in a string.

Explanation:
Vowels: a, e, i, o, u (consider both lower and upper case).
We:
1) Convert string to lower-case for simplicity.
2) Loop through each character and check if it is in the set of vowels.
3) Maintain a counter and print the result.

```python
text = input('Enter any string:')

vowels = 'aeiouAEIOU'

count = 0

for ch in text:
    if ch in vowels:
        count += 1
print(f'Total vowels in string are: {count}')
```

### Q21: Write a Python program to reverse a string.

Explanation:     <br>
Strings support slicing.     <br>
We can use s[::-1] to reverse the full string.    <br>
Alternatively, we can build the reversed string character by character, but slicing is simpler.


```python
str1 = input('Enter string:')
rev_str = str1[::-1]
print(f'Reversed string: {rev_str}')
```

### Q22: Write a Python program to check if a string is a palindrome.

Explanation:  <br>
A palindrome reads the same forwards and backwards.    <br>
We can reverse the string using slicing and compare it with the original.   <br>
If both are equal, it is a palindrome.

```python
str1 = input('Enter any string:')
rev_str = str1[::-1]
if str1 == rev_str:
    print(f'{str1} is palindrome')
else:
    print(f'Given string is not palindrome')
```
