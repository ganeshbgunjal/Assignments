### Q2: Write a Python program to calculate the factorial of a number using a loop.

Explanation:
The factorial of a number n (written as n!) is the product of all positive integers from 1 to n.
Example: 5! = 1 * 2 * 3 * 4 * 5 = 120.
We can use a for loop from 1 to n, multiply each value, and store the result in a variable.

```python
num = int(input('Enter number: '))

factorial = 1
for i in range(1,num+1):
    factorial = factorial*i
print(f'Factorial of {num} is {factorial}')
```

### Q3: Write a Python program to find the sum of the first n natural numbers.

Explanation:
Natural numbers start from 1: 1, 2, 3, ...
To find sum of first n natural numbers, we can:
1) Use a loop to add numbers from 1 to n.
2) Or use the formula n * (n + 1) // 2.
Here we will use a loop for better practice.

```python
num = int(input('Enter number:'))
sum_n = 0

for i in range(1,num+1):
    sum_n+=i
print(f'Sum of first {num} numbers is {sum_n}')
```

### Q4: Write a Python program to check whether a given year is a leap year.

Explanation:
A leap year:
- is divisible by 400, OR
- is divisible by 4 but NOT divisible by 100.
We check these conditions using if-else and the modulo (%) operator.

```python
year= int(input('Enter year:'))
if (year%4==0 and year%100!=0) or year%400==0:
    print('Year is leap year')
else:
    print('Year is not leap year')
```

### Q5: Write a Python program to sort a list of numbers in ascending order.

Explanation:
Python provides:
- sorted(list) → returns a new sorted list.
- list.sort() → sorts the list in-place.
Here, we will show both methods with a simple example list.


```python
lst = [1,8,3,9,5,2,7,5,8]

sorted_list = sorted(lst)
print(sorted_list)
```

```python
nums = [5, 3, 9, 1, 4]

nums.sort()
print("Sorted in-place:", nums)
```

### Q6: Write a Python program to check if a number is even or odd.

Explanation: <br>
A number is even if it is divisible by 2 (remainder 0 when divided by 2).   <br>
We use the modulo operator % to check n % 2.    <br>
If n % 2 == 0 → even, else → odd.

```python
num = int(input('Enter number: '))
if num % 2 == 0:
    print(f'{num} is Even')
else:
    print(f'{num} is Odd')
```
