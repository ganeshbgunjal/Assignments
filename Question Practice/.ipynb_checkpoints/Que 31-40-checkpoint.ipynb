### Q32: Write a Python program to find the maximum element in a list without using max().

Explanation:   <br>
We initialize maximum with the first element of the list.   <br>
We loop through the list, and if any element is greater than maximum, we update maximum.  <br>
Finally, we print the maximum value.

```python
lst = [10,20,60,30,40,50]
max_element = lst[0]

for num in lst:
    if num > max_element:
        max_element = num
print(f'Max element in the list: {max_element}')


        
```

### Q33: Write a Python program to remove duplicates from a list while keeping the original order.

Explanation:   <br>
We iterate over the list and maintain a new list called result.    <br>
For each element, if it is not already in result, we append it.    <br>
This way we keep only first occurrence of each element and preserve order.

```python
nums = [1, 2, 2, 3, 4, 4, 5, 1]
result = []

for i in nums:
    if i not in result:
        result.append(i)
print(f'List without duplicates: {result}')
```

### Q34: Write a Python program to reverse a list without using reverse() or slicing.

Explanation:  <br>
We use two indexes: i at the start and j at the end.    <br>
While i < j, we swap elements at i and j.    <br>
Then we move i forward and j backward until they meet.

```python
numbers = [10, 20, 30, 40]
rev_lst = []

for i in range(len(numbers)-1,-1,-1):
    rev_lst.append(numbers[i])

print(f'Reversed List: {rev_lst}')
```

```python
nums = [1, 2, 3, 4, 5]
i = 0
j = len(nums) - 1

while i < j:
    nums[i], nums[j] = nums[j], nums[i]
    i += 1
    j -= 1

print("Reversed list:", nums)
```

### Q35: Write a Python program to find the second largest element in a list.

Explanation:   <br>
We can remove duplicates using set(), then sort the unique values.    <br>
If at least two unique elements exist, the second largest is at index -2 of the sorted list.   <br>
If not, we print an appropriate message.

```python
nums = [10, 20, 4, 45, 99, 99]

unique_nums = list(set(nums))
print(unique_nums)

if len(unique_nums)<2:
    print('Not enough numbers to find second largest number.')
else:
    unique_nums.sort()
    second_largest = unique_nums[-2]
    print(f'Second largest num: {second_largest}')
```

### Q36: Write a Python program to count occurrences of each element in a list.

Explanation:    <br>
We use a dictionary to store frequencies.   <br>
We loop over the list; for each element, we increment its count in the dictionary.   <br>
At the end, we print each element and its frequency.

```python
nums = [1, 2, 2, 3, 1, 4, 2,10,10,4]
freq = {}

for num in nums:
    if num in freq:
        freq[num]+=1
    else:
        freq[num] = 1

for key, value in freq.items():
    print(key,':',value)
```

### 37: Write a Python program to create a dictionary from two lists: one of keys and one of values.

Explanation:    <br>
We assume both lists are of the same length.   <br>
We can loop using index or use zip() to pair keys and values.    <br>
Then we insert each pair into a dictionary.

```python
keys = ['name', 'age', 'city']
values = ['Ganesh', 33, 'Pune']

result = dict(zip(keys,values))
print(result)
```

### Q39: Write a Python program to find the intersection of two sets.

Explanation:    <br>
Sets support mathematical operations like intersection using & or the intersection() method.  <br>
We define two sets, then compute their intersection and print it.

```python
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7}

intersection = set1.intersection(set2)

print(f'Intersection of 2 sets: {intersection}')
```

### Q40: Write a Python program to find the union and difference of two sets.

Explanation:    <br>
Union contains all unique elements from both sets, using | or union().   <br>
Difference A - B contains elements present in A but not in B.   <br>
We compute and print both union and differences.

```python
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7}

union = set1.union(set2)
difference = set1.difference(set2)

print(f'Union of 2 sets: {union}')
print(f'Difference of 2 sets: {difference}')
```
