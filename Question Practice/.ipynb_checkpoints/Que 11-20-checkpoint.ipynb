### Q11: Write a Python program to print numbers from 1 to 10 using a while loop.

Explanation:
Initialize i = 1.
Use while i <= 10:
- print i
- increment i by 1 each time.
Loop stops when condition becomes false.

```python
i=1
while i <=10:
    print(i)
    i = i+1
    
```

### Q12: Write a Python program to print the multiplication table of a given number n (from 1 to 10).

Explanation:    <br>
Take an integer n.      <br>
Use a for loop from 1 to 10.      <br>
On each iteration, multiply n by i and print “n x i = result”.

```python
num = int(input('Enter num:'))
for i in range(1,11):
    print(f'{num} * {i} = {i*num}')
```

### Q13: Write a Python program to calculate the sum of all even numbers from 1 to n.

Explanation:
We want to add only even numbers.
Approach:
- Loop i from 1 to n.
- If i % 2 == 0, add to sum.
Finally print the sum.

```python
num = int(input('Enter number:'))
total = 0
for i in range(1,num+1):
    if i%2==0:
        total +=i
print(total)
        
```

### Q14: Write a Python program to count how many numbers between 1 and 100 are divisible by 3.

Explanation:    <br>
We loop from 1 to 100.     <br>
If a number i satisfies i % 3 == 0, we increment a counter.     <br>
At the end, we print the count.

```python
counter = 0
for i in range(1,101):
    if i%3==0:
        counter+=1
print(counter)
```

### Q17: Write a Python program to find the factorial of a number using recursion.

Explanation:
Recursive definition of factorial:
- factorial(0) = 1
- factorial(1) = 1
- factorial(n) = n * factorial(n - 1) for n > 1
We call the function recursively until the base case.

```python
num = int(input('Enter num:'))
def factorial(num):
    if num ==0 or num==1:
        return 1
    else:
        return num * factorial(num-1)
print(f'Factorial of {num} is {factorial(num)}')
```

### Q16: Write a Python program to reverse the digits of a number using a loop.

Explanation:
We can repeatedly take the last digit using n % 10, and build a new number.
Steps:
1) Initialize rev = 0.
2) While n > 0:
   - digit = n % 10
   - rev = rev * 10 + digit
   - n = n // 10 (integer division)
3) rev will be the reversed number.

```python
num = input('Enter num:')
rev= ""

for digit in num:
    rev = digit + rev
print(f'Reversed: {rev}')
```

```python
num = int(input("Enter a number: "))
rev = 0

while num > 0:
    digit = num % 10        # extract last digit
    rev = rev * 10 + digit  # build reversed number
    num = num // 10         # remove last digit

print("Reversed number:", rev)
```

```python
num = int(input('Enter num:'))
rev = 0

while num >0:
    digit = num %10    #get last digit
    rev = rev * 10 + digit    #build reversed number
    num = num // 10    #remove last digit
print(f'Reversed Number: {rev}')
    
```
