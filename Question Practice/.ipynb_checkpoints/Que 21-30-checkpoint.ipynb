### Q21: Write a Python program to count the number of vowels in a string.

Explanation:
Vowels: a, e, i, o, u (consider both lower and upper case).
We:
1) Convert string to lower-case for simplicity.
2) Loop through each character and check if it is in the set of vowels.
3) Maintain a counter and print the result.

```python
text = input('Enter any string:')

vowels = 'aeiouAEIOU'

count = 0

for ch in text:
    if ch in vowels:
        count += 1
print(f'Total vowels in string are: {count}')
```

### Q21: Write a Python program to reverse a string.

Explanation:     <br>
Strings support slicing.     <br>
We can use s[::-1] to reverse the full string.    <br>
Alternatively, we can build the reversed string character by character, but slicing is simpler.


```python
str1 = input('Enter string:')
rev_str = str1[::-1]
print(f'Reversed string: {rev_str}')
```

### Q22: Write a Python program to check if a string is a palindrome.

Explanation:  <br>
A palindrome reads the same forwards and backwards.    <br>
We can reverse the string using slicing and compare it with the original.   <br>
If both are equal, it is a palindrome.

```python
str1 = input('Enter any string:')
rev_str = str1[::-1]
if str1 == rev_str:
    print(f'{str1} is palindrome')
else:
    print(f'Given string is not palindrome')
```

### Q23: Write a Python program to split a sentence into words and print each word on a new line.

Explanation:  <br>
We can use the split() method, which by default splits on spaces.

Then we loop through the list of words and print each word separately.

```python
word = input('Enter any word:')
words = word.split()

print('Words---')

for w in words:
    print(w)
```

### Q24: Write a Python program to find the longest word in a sentence.

Explanation:
Steps:
1) Split the sentence into words using split().
2) Keep a variable longest initialized to empty string.
3) For each word, if its length is greater than current longest, update longest.
4) Print the longest word.

```python
sentence = input('Enter sentence:')
words = sentence.split()

longest_word = ''
for word in words:
    if len(word) > len(longest_word):
        longest_word= word
print(f'longest word: {longest_word}')
```

### 26: Write a Python program to count the frequency of each character in a string.

Explanation:    <br>
We can use a dictionary to store counts.   <br>
For each character:    <br>
- If it's already a key in dictionary, increment its count.  <br>
- Else, add it with count 1.   <br>
At the end, print the dictionary.

```python
str1 = input('Enter the string:')
freq = {}

for ch in str1:
    if ch in freq:
        freq[ch]+=1
    else:
        freq[ch] =1
print('Character frequencies:--')
for ch,count in freq.items():
    print(ch, ":", count)
```

### Q27: Write a Python program to create a list of 5 numbers and print their sum and average.

Explanation:   <br>
We define a list of numbers.   <br>
We can use sum(list) to get total, and len(list) to get count.   <br>
Average = total / count.

```python
nums = [10, 20, 30, 40, 50]
total = sum(nums)

average = total/len(nums)

print(f'Total:{total}')
print(f'Average:{average}')
```

### Q28: Write a Python program to find the maximum element in a list without using max().

Explanation:
We:
1) Assume first element is the current maximum.
2) Loop over the list.
3) If we find an element greater than the current maximum, we update it.
At the end, we print the maximum.

```python
nums = [3, 7, 1, 9, 4,33,76]

maximum_element = nums[0]

for i in nums:
    if i >maximum_element:
        maximum_element = i
print(f'Maximum element is {maximum_element}')
```

### Q29: Write a Python program to remove duplicates from a list while keeping the original order.

Explanation:  <br>
We iterate through the list and maintain a second list result.
For each element:
- If it is not already in result, append it.
This keeps only the first occurrence of each element.

```python
nums = [1, 2, 2, 3, 4, 4, 5, 1]

result = []
for i in nums:
    if i not in result:
        result.append(i)

print(f'Original list: {nums}')
print(f'List without duplicates: {result}')
    
```
